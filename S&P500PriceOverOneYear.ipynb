Import appropriate packages

```python
import pandas as pd #library to read the csv file
import matplotlib.pyplot as plt #graphing library
plt.style.use('fivethirtyeight') # style of graph
%matplotlib inline
#based on my research %matplot inline,is used to geneerate the visual on jupyter notebook
```

Function to read and create graph and assign x and y

```python
spxdata = pd.read_excel('C:\\Users\\Andrew\\Documents\\Python Scripts\\SPX1Year.xlsx')#csv file of S&P 500 dates and prices
x = spxdata['Date'] #x axis is the date and moves over time along the x axis
y = spxdata['Close'] #y axis is the closing price and goes higher or lower
    
plt.plot(x,y)#plotting the Date and Time varibles
plt.xlabel('Date')#name of x label
plt.ylabel('Closing Price (US Dollars)')#name of y label
plt.title('S&P 500 Price Movement Over The Past Year')# name of graph
```

```python
plt.show()#generates the graph
```
